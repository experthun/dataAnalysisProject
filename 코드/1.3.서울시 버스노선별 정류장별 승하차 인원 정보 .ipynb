### 3. 버스 정류장별 일평균

```python
df_g_sum = df_b.groupby('버스정류장ARS번호', as_index = False).mean() 
```

```python
#df_g_sum['승차총승객수'] = df_g_sum['승차총승객수'] / 365
#df_g_sum['하차총승객수'] = df_g_sum['하차총승객수'] / 365
df_g_sum
```

```python
df_g_sum['승하차평균'] = df_g_sum['승차총승객수'] + df_g_sum['하차총승객수'] 
df_g_sum
```

```python
#에러값 제거
df_g_sum = df_g_sum.drop(12768)
```

```python
df_g_sum.tail()
```

```python
#버스정류장ARS번호 자료형을 숫자형으로 변환
df_g_sum['버스정류장ARS번호']= pd.to_numeric(df_g_sum['버스정류장ARS번호'])
```

```python
df_g_sum.info()
```

```python
#승차총승객수 컬럼을 기준으로 데이터를 정렬
df_g_sum.sort_values(by = ['승차총승객수'])
```

### 5. 데이터 병합

```python
#버스정류장ARS번호를 ARS-ID로 일치 시킴
df_g_sum.rename(columns={'버스정류장ARS번호':'ARS-ID'}, inplace = True)
df_g_sum.head(2)
```

```python
df_merge = pd.merge(df_g_sum, df_loc, on = 'ARS-ID') #df_g_sum(버스 승하차 인원) 데이터와 df_loc(서울시 정류소 리스트) 데이터를 ARS-ID 기준으로 병합
df_merge
```

```python
df_merge = df_merge[['ARS-ID', '승차총승객수', '하차총승객수', '승하차평균', '정류소명', '좌표X', '좌표Y']]
df_merge.head(1)
```
