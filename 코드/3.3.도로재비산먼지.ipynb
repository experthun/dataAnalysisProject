# 도로 재비산먼지 데이터 전처리

### 1. 데이터 불러오기

```python
import pandas as pd
import os

file_path = os.listdir()
file_list = [file for file in file_path if file.endswith(".csv")]

#파일 목록 확인
print(file_list)
print(len(file_list))
```

```python
#도로 재비산먼지 측정정보 2021년 1월 ~ 12월 데이터 불러오기.
fug_1 = pd.read_csv('한국환경공단_도로 재비산먼지 측정정보_20210131_cng.csv', encoding = 'cp949')
fug_2 = pd.read_csv('한국환경공단_도로 재비산먼지 측정정보_20210228_cng.csv', encoding = 'cp949')
fug_3 = pd.read_csv('한국환경공단_도로 재비산먼지 측정정보_20210331_cng.csv', encoding = 'cp949')
fug_4 = pd.read_csv('한국환경공단_도로 재비산먼지 측정정보_20210430_cng.csv', encoding = 'cp949')
fug_5 = pd.read_csv('한국환경공단_도로 재비산먼지 측정정보_20210531_cng.csv', encoding = 'cp949')
fug_6 = pd.read_csv('한국환경공단_도로 재비산먼지 측정정보_20210630_cng.csv', encoding = 'cp949')
fug_7 = pd.read_csv('한국환경공단_도로 재비산먼지 측정정보_20210731_cng.csv', encoding = 'cp949')
fug_8 = pd.read_csv('한국환경공단_도로 재비산먼지 측정정보_20210831_cng.csv', encoding = 'cp949')
fug_9 = pd.read_csv('한국환경공단_도로 재비산먼지 측정 정보_20210930_cng.csv', encoding = 'cp949')
fug_10= pd.read_csv('한국환경공단_도로 재비산먼지 측정 정보_20211031_cng.csv', encoding = 'cp949')
fug_11= pd.read_csv('한국환경공단_도로 재비산먼지 측정 정보_20211130_cng.csv', encoding = 'cp949')
fug_12= pd.read_csv('한국환경공단_도로 재비산먼지 측정 정보_20211231_cng.csv', encoding = 'cp949')
```

```python
fug_1.info() #데이터 정보 확인
```

```python
#도로 재비산먼지 측정정보 2021년 1월부터 ~ 12월까지 데이터 결합
fug = pd.concat([fug_1, fug_2, fug_3, fug_4, fug_5, fug_6, fug_7, fug_8, fug_9, fug_10, fug_11, fug_12],ignore_index = True)
```

```python
fug.info() #결합한 데이터 출력 
```

```python
fug.sample(10) #데이터가 잘 결합되었는지 확인하기 위해 랜덤으로 10개의 데이터 출력
```

### 2. 필요한 지역 추출

```python
fug['지역'].value_counts()  #지역별 데이터  총 개수 확인
```

```python
fug = fug[fug['지역'] == '서울'] #서울지역 데이터만 추출
```

```python
#fug.drop(['기온 (℃)','습도 (%)','기온','습도'], axis='columns', inplace = True) 
#fug = fug.drop(['기온 (℃)','습도 (%)','기온','습도'], axis='columns') 
fug.drop(['기온 (℃)','습도 (%)','기온','습도'], axis='columns', inplace = True) # 데이터 분석시 불필요한 컬럼 제거
```

### 3. 평균 계산

```python
fug_mean = fug.groupby(['지역명','도로명'], as_index = False).mean() #서울시 지역명 도로명을 그룹으로 묶어 재비산먼지 평균농도 출력
```
